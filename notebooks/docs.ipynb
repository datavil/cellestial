```python
import cellestial as cl
import scanpy as sc

from lets_plot import *

LetsPlot.setup_html()

data = sc.read_h5ad("data/pbmc3k_pped.h5ad")
```

```python
import cellestial as cl
import scanpy as sc

from lets_plot import *

LetsPlot.setup_html()

violin = (
    cl.violin(
        data,
        "CD14",
        fill="cell_type_lvl1",
        scale="width",
        point_size=2,
        threshold=0.1,
        trim=False,
    )
    + ggsize(800, 400)
    + scale_fill_brewer(palette="Set2")
    + guides(fill=guide_legend(ncol=2))
)

violin
```
