```python
all_markers = set()
for markers in marker_genes.values():
    all_markers.update(markers)
all_markers = list(all_markers)
```

```python
dot = (
    cl.dotplot(
        data,
        keys=all_markers,
        group_by="cell_type_lvl1",
        color_high="red",
        #shape=21,
        #fill=True,
        stroke=0.1,
        #color="#1f1f1f",
        sort_by="avg_exp",
        tooltips=layer_tooltips(["gene","pct_exp","avg_exp","cell_type_lvl1"])
    )
    + scale_size(trans="sqrt", breaks=[0, 25, 50, 75, 100])
    + scale_y_discrete(expand=[0.1, 0.1])
    + theme(axis_text_y=element_text(angle=90,size=12))
)
```
