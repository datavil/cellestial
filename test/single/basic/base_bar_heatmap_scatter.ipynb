```python
import scanpy as sc

import cellestial as cl

data = sc.read("data/pbmc3k_pped.h5ad")
from lets_plot import *

LetsPlot.setup_html()
```

```python
data.obs
```

## Base Plot

```python

(
    cl.plot(data, aes("cell_type_lvl1", "n_genes"))
    + geom_violin(aes(fill="cell_type_lvl1"), scale="width")
    + geom_boxplot(width=0.2,outlier_size=0)
    + scale_fill_viridis()
)
```

## Barplot

```python
cl.bar(data, aes("cell_type_lvl1", fill="leiden")) + scale_fill_brewer(palette="Set2")
```

```python
cl.bar(
    data,
    aes("leiden", fill="predicted_doublet"),
) + scale_fill_brewer(palette="Set2", direction=-1)
```

## Scatter

```python
cl.scatter(
    data, aes(x="n_genes_by_counts", y="total_counts", color="cell_type_lvl1"), alpha=0.6
) + geom_smooth(aes(x="n_genes_by_counts", y="total_counts", color="cell_type_lvl1"))
```
