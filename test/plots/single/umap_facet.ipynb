```python
import cellestial as cl
import scanpy as sc
from lets_plot import *

LetsPlot.setup_html()

data = sc.read("data/pbmc3k_pped.h5ad")
```

```python
data.obs
```

```python
dim = cl.dimensional(
    data,
    dimensions="umap",
    key="leiden",
    size=1,
    axis_type=None,
    add_tooltips=["n_genes", "total_counts_hb", "MT-ND2"],
    legend_ondata=True,
    ondata_size=12,
    ondata_color="black",
    ondata_fontface="bold",
    ondata_family="sans",
    ondata_alpha=0.8,
)
dim + ggsize(800, 600)
```

```python
(
    dim
    + facet_wrap(facets=["leiden"])
    + ggsize(800, 600)
)
```
