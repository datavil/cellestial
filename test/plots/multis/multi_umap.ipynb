```python
import cellestial as cl
import scanpy as sc

from lets_plot import *

data = sc.read("data/pbmc3k_mini.h5ad")
```

```python
data.var_names
```

```python
cl.umaps(
    data,
    axis_type="arrow",
    keys=["leiden", "HBD", "NEAT1", "IGKC"],
    ncol=2,
    size=1,
    layers=[theme(legend_title=element_text(size=12))],
) + ggsize(900, 700)
```

```python
(
    cl.umaps(
        data,
        keys=["leiden", "HBD", "NEAT1", "IGKC"],
        ncol=2,
        size=1,
        color_high="red",
    )
    + ggsize(900, 700)
).to_png("assets/multi_umap.png")
```

```python
sc.pp.highly_variable_genes(data, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
gggrid(
    [
        cl.umap(data),
        cl.umap(data, key="HBD"),
        cl.umap(data, key="NEAT1"),
        cl.umap(data, key="IGKC"),
    ],
    ncol=2,
    sharex="all",
    sharey=True,
)
```

```python
cl.umap(data, key="CHD5", size=2, axis_type=None) + ggsize(700, 600)
```

```python
plot1 = ggplot({"x": [1, 2, 3], "y": [4, 5, 6]}) + geom_point(aes("x", "y"))
plot2 = ggplot({"x": [2, 3, 4], "y": [6, 7, 8]}) + geom_point(aes("x", "y"))
plot3 = ggplot({"x": [1, 2, 3], "y": [4, 5, 6]}) + geom_point(aes("x", "y"))
plot4 = ggplot({"x": [2, 3, 4], "y": [6, 7, 8]}) + geom_point(aes("x", "y"))

# Combine plots in a grid with shared axes
grid = gggrid([plot1, plot2, plot3, plot4], ncol=2, sharex="all", sharey="all")


grid = (
    grid
    + theme_classic()
    + ggsize(600, 400)
    + theme(axis_title=element_text(size=20, face="bold"))
)
grid.to_png("issue.png")
grid += theme(plot_title="Overall X-Axis Label")
grid
```
