```python
all_markers = set()
for markers in marker_genes.values():
    all_markers.update(markers)

all_markers = list(all_markers)
```

```python
all_markers
```

```python
vln = cl.violins(
    data,
    all_markers,
    # color="cell_type_lvl1",
    fill="cell_type_lvl1",
    #scale="width",
    trim=False,
    show_points=False,
    ncol=1,
    sharex=True,
    sharey=True,
    guides="collect",
    vspace=0,
    hspace=0,
    share_ticks=True,
    size=0.1,
    scale="width",
)
vln
```
